```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mdates
import numpy as np
import matplotlib.ticker as ticker
from datetime import datetime, timedelta
```

```python
data = pd.read_csv ('/media/naureeng/Data/AAPL.csv')
x = data['Date']
y = data['Close']
ax = plt.axes()
ax.plot(x,y, linestyle='solid')

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # use maximum of 3 ticks 
#ax.xaxis.set_major_formatter(plt.NullFormatter()) use no ticks 

plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['axes.linewidth'] = 3
plt.tight_layout()
```

```python
ax = plt.axes()
ax.plot(x,y, linestyle='solid')

loc = ticker.MultipleLocator(base=5.0) # use every 5 ticks 
ax.xaxis.set_major_locator(loc)

plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['axes.linewidth'] = 3
plt.tight_layout()
```
